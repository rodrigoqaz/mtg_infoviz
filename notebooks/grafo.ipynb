## Verificar sinergia das cartas

```python
import networkx as nx
import pandas as pd
from pyvis.network import Network
from pyedhrec import EDHRec

# Inicializar EDHRec
edhrec = EDHRec()

# Comandante
cmd = "Ayara, First of Locthwain"

# Obter as cartas associadas ao comandante
miirym_cards = edhrec.get_commander_cards(cmd)

# Inicializar listas finais
cards_with_synergy = []
cards_without_synergy = []

# Verificar sinergia e preencher listas finais
sections = ['New Cards', 'High Synergy Cards', 'Top Cards', 'Creatures', 'Instants', 'Sorceries', 'Utility Artifacts', 'Enchantments', 'Planeswalkers', 'Utility Lands', 'Mana Artifacts', 'Lands']
for card in cards:
    found = False
    for section in sections:
        if section in miirym_cards:
            for new_card in miirym_cards[section]:
                if new_card['name'].lower() == card.lower():
                    cards_with_synergy.append({'name': card, 'synergy': new_card['synergy']})
                    found = True
                    break
        if found:
            break
    if not found:
        cards_without_synergy.append(card)

# Resultados
print("Cartas com sinergia:")
for card in cards_with_synergy:
    print(f"Nome: {card['name']}, Sinergia: {card['synergy']}")

print("\nCartas sem sinergia:")
for card in cards_without_synergy:
    print(card)
```

## construir o GRafo das cartas com sinergia

```python
import requests
import networkx as nx
from pyvis.network import Network


def get_card_image_url(card_name):
    response = requests.get(f"https://api.scryfall.com/cards/named?exact={card_name}")
    if response.status_code == 200:
        card_data = response.json()
        if 'image_uris' in card_data:
            return card_data['image_uris']['normal']
        else:
            # Some cards may not have 'image_uris', check for other types or return a placeholder
            return card_data['card_faces'][0]['image_uris']['normal'] if 'card_faces' in card_data else ""
    else:
        print(f"Erro ao obter imagem da carta {card_name}")
        return ""

# Adiciona URLs de imagens às cartas com sinergia
for card in cards_with_synergy:
    card['image_url'] = get_card_image_url(card['name'])

# Adiciona URLs de imagens às cartas sem sinergia
cards_without_synergy_with_images = [{'name': card, 'image_url': get_card_image_url(card)} for card in cards_without_synergy]

# Inicializa os grafos
G_with_synergy = nx.Graph()
G_without_synergy = nx.Graph()

# Adiciona o nó do comandante para ambos os grafos
commander = 'Ayara, First of Locthwain'
commander_image = get_card_image_url(commander)
G_with_synergy.add_node(commander, label=commander, color='red', size=20, image=commander_image, shape='image')
G_without_synergy.add_node(commander, label=commander, color='red', size=20, image=commander_image, shape='image')

# Adiciona os nós das cartas com sinergia
for card in cards_with_synergy:
    color = 'green'
    G_with_synergy.add_node(card['name'], label=f"{card['name']} (synergy: {card['synergy']})", color=color, size=10 + card['synergy']*20, image=card['image_url'], shape='image')
    G_with_synergy.add_edge(commander, card['name'], weight=card['synergy'])

# Adiciona os nós das cartas sem sinergia
for card in cards_without_synergy_with_images:
    color = 'gray'
    G_without_synergy.add_node(card['name'], label=card['name'], color=color, size=10, image=card['image_url'], shape='image')
    G_without_synergy.add_edge(commander, card['name'], weight=1)

# Cria o grafo interativo com a biblioteca pyvis
nt_with_synergy = Network(notebook=True, height='750px', width='100%', bgcolor='white', font_color='black')
nt_with_synergy.from_nx(G_with_synergy)
nt_with_synergy.show('deck_with_synergy.html')

# nt_without_synergy = Network(notebook=True, height='750px', width='100%', bgcolor='white', font_color='black')
# nt_without_synergy.from_nx(G_without_synergy)
# nt_without_synergy.show('deck_without_synergy.html')
```

```python
import requests
import networkx as nx
from pyvis.network import Network

# Função para obter a URL da imagem da carta
def get_card_image_url(card_name):
    response = requests.get(f"https://api.scryfall.com/cards/named?exact={card_name}")
    if response.status_code == 200:
        card_data = response.json()
        if 'image_uris' in card_data:
            return card_data['image_uris']['normal']
        else:
            # Algumas cartas podem não ter 'image_uris', verificar outros tipos ou retornar um placeholder
            return card_data['card_faces'][0]['image_uris']['normal'] if 'card_faces' in card_data else ""
    else:
        print(f"Erro ao obter imagem da carta {card_name}")
        return ""

# Função para sanitizar o nome da carta
def sanitize_card_name(card_name):
    return card_name.lower().replace(' ', '-').replace(',', '').replace("'", "").strip()

# Lista de cartas do deck

# Função para ler o arquivo de combos
def read_combos_from_file(file_path):
    with open(file_path, 'r') as file:
        import json
        combos_data = json.load(file)
    return combos_data

# Carregar os combos do arquivo

combos_base=cmd_combos
deck_cards=cards
if cmd not in deck_cards:
    deck_cards.append(cmd)


# Função para verificar combos
def check_combos(deck, combos):
    sanitized_deck = {sanitize_card_name(card): card for card in deck}
    result = []
    for combo in combos['container']['json_dict']['cardlists']:
        combo_cards = [(card['name'], card['sanitized']) for card in combo['cardviews']]
        cards_in_deck = [original_name for original_name, sanitized_name in combo_cards if sanitized_name in sanitized_deck]
        cards_not_in_deck = [original_name for original_name, sanitized_name in combo_cards if sanitized_name not in sanitized_deck]
        result.append({
            'combo': combo['tag'],
            'cards': [original_name for original_name, sanitized_name in combo_cards],
            'cards_in_deck': cards_in_deck,
            'cards_not_in_deck': cards_not_in_deck
        })
    return result

# Verifica os combos
combo_results = check_combos(deck_cards, combos_base)

# Inicializa o grafo
G = nx.Graph()

# Adiciona nós e arestas para cada combo
for result in combo_results:
    for card in result['cards']:
        image_url = get_card_image_url(card)
        if card not in G:
            G.add_node(card, label=card, image=image_url, shape='image')
        if card in result['cards_in_deck']:
            G.nodes[card]['color'] = 'green'
        else:
            G.nodes[card]['color'] = 'red'
        
        # Adiciona informações sobre o combo
        if 'title' in G.nodes[card]:
            G.nodes[card]['title'] += f"<br>{result['combo']}"
        else:
            G.nodes[card]['title'] = result['combo']

    # Adiciona arestas entre as cartas do combo
    for i in range(len(result['cards']) - 1):
        G.add_edge(result['cards'][i], result['cards'][i+1])

# Cria o grafo interativo com a biblioteca pyvis
net = Network(notebook=True, height='750px', width='100%', bgcolor='#ffffff', font_color='black')
net.from_nx(G)
net.show('combos_graph.html')
```
